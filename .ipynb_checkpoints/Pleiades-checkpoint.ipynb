```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

stardata = pd.read_csv("stardata.csv", header = None, sep = ',')
stardata.columns = ["RA", "Decl", "Parallax", "AppMagn"]
```

```python
stardata_ = stardata.query('Parallax != "--"').reset_index()
AbsMagn = np.zeros(len(stardata_))

for i in range(len(stardata_)):
    AbsMagn[i] = float(stardata_['AppMagn'][i]) + 5 + 5*np.log(1000/float(stardata_['Parallax'][i]))/np.log(10)
    
plt.hist(AbsMagn,50)
plt.xlabel('AbsMagn')
plt.ylabel('No. of stars')
plt.title('Absolute Magnitude Histogram')
```

```python
plt.scatter(stardata_['RA'], stardata_['Decl'], (12-stardata_['AppMagn'])*3)
```
