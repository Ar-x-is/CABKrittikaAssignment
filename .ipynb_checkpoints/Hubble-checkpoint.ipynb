```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
data_ = pd.read_csv("data.txt", sep=',')
data = data_
data['mod0'] = 10**((data['mod0']/5)-5)
data.columns = ['d','vgsr']

d_ = data['d']
vgsr_ = data['vgsr']
```

```python
def linFunc(x, m, c):
    return c - m*x
    
[m_, c_], p_cov = curve_fit(linFunc, d_, vgsr_)
print(m_, c_)

plt.scatter(d_, vgsr_, 0.5)
plt.plot(d_, linFunc(d_, m_, c_), 'r--', linewidth=1)
```

```python
# Data Cleaning
delta = data['vgsr'] - linFunc(d_, m_, c_)
sigma = np.std(delta)

# If the vgrs value is within 3 standard deviations of the best fit
dataFiltered = data[abs(delta) <= 3*sigma]
print(dataFiltered)
dfiltered = dataFiltered['d']
vgsrfiltered = dataFiltered['vgsr']

[m, c], p_cov = curve_fit(linFunc, dfiltered, vgsrfiltered)
print(m, c)

plt.scatter(dfiltered, vgsrfiltered, 0.5)
plt.plot(dfiltered, linFunc(dfiltered, m, c), 'r--', linewidth=1)
```
